# Определение неэффективных операторов.

«Нупозвони» — провайдер виртуальный телефонии.  Его клиенты — колл-центры, которые:

* распределяют входящие вызовы на операторов,
* совершают исходящие вызовы силами операторов.

Необходимо найти самых неэффективных операторов. Признаки низкой
эффективности:

* много пропущенных входящих вызовов,
* долгое ожидания ответа при входящих внешних вызовах,
* мало исходящих вывозов — в случае тех колл-центров, что специализируются на обзвонах, а не обработке входящих звонков.

## Расчеты и добавление данных в таблицу

### Длительность ожидания звонка

Посчитаем среднюю длительность ожидания звонка для каждого оператора на определенную дату. Запишем данные в новый столбец `waiting_duration_mean`.

```python
data['waiting_duration_mean'] = (data['total_call_duration'] - data['call_duration']) / data['calls_count']
```

###  Длительность одного звонка с учетом времени ожидания ответа.

Добавим информацию о длительности одного звонка с учетом времени ожидания ответа. Сохраним в столбце `total_one_call_duration`.

```python
data['total_one_call_duration'] = data['total_call_duration'] / data['calls_count']
```

### Длительность одного звонка без учета времени ожидания ответа.

Добавим информацию о длительности одного звонка без учета времени ожидания ответа. Сохраним в столбце `one_call_duration`.

```python
data['one_call_duration'] = data['call_duration'] / data['calls_count']
```

### Категоризация оператора.

Для каждого оператора посчитаем, сколько приходится на него входящих и исходящих вызовов.

```python
# сгруппируем данные по операторам и посчитаем суммарное количество входящих и исходящих звонков
data_pivot = (data
              .pivot_table(index='operator_id', columns='direction', values='calls_count', aggfunc='sum')
              .reset_index()
             )
# переименуем колонки
data_pivot.columns=['operator_id', 'in_total', 'out_total']
# заменим пустые значения в колонках "in_total"  и "out_total" на ноль
data_pivot['in_total'] = data_pivot['in_total'].fillna(0)
data_pivot['out_total'] = data_pivot['out_total'].fillna(0)
display(data_pivot.head())
```

Так как у операторов так же возможны исходящие внутренние вызовы, то установим порог в 70% вызовов от общего числа, при котором будем относить оператора к той или иной категории. 

Если более 70% звонков иходящие - отнесем к категории `out`, если более 70% звонков входящие, то отнесем к категории `in`. Во всех остальных случаях будем считать, что оператор является многофункциональным, т.е. совершает и входящие и исходящие вызовы, отнесем его к категории `mixed`.

```python
# создадим функцию category_operator(row), которая относит оператора к определенной категории 
def category_operator(row): 
    if row['in_total'] > 0.7 * (row['in_total'] + row['out_total']):
        return 'in'
    if  row['out_total'] > 0.7 * (row['in_total'] + row['out_total']):
        return 'out'
    else:
        return 'mixed'
# создадим новый столбец 'operator_type', в котором сохраним категорию оператора  
data_pivot['operator_type'] = data_pivot.apply(category_operator, axis=1)  
```

Посчитаем сколько получилось операторов каждого типа. Выведем 5 первых строк сводной таблицы `data_pivot`.

```python
display(data_pivot.head())
data_pivot['operator_type'].value_counts()
```

Выделив категории операторов мы получили:
* 635 операторов, которые специализируются на исходящих вызовах;
* 288 операторов специализируется на входящих звонках;
* 169 операторов работают в двух направлениях.

Построим гистограмму количества разных типов операторов.

```python
plt.figure(figsize=(10, 7))
(data_pivot['operator_type']
 .value_counts()
 .plot(kind='bar', grid=True)
)
plt.title('Распределение категорий операторов', fontsize=15)
plt.xlabel('Направление звонка', fontsize=15)
plt.ylabel('Количество', fontsize=15);
```

Объеденим таблицы `data_pivot` и `data` по столбцу `operator_id`.

```python
data = data.merge(data_pivot, on='operator_id', how='left')
```

## Проверка гипотез

### гипотеза первая

Среднее время ожидания при входящих внешних вызовах в группах `in` и `mixed` одинаковое.

Выдвенем статистические гипотезы:
* H0 - среднее время ожидания при входящих внешних вызовах в группах `in` и `mixed` равно;
* H1 - среднее время ожидания при входящих внешних вызовах в группах `in` и `mixed` не равно.

Используем метод `scipy.stats.ttest_ind()`, так как мы проверяем гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.

```python
# проведем t-test
alpha = .01 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    data_in_internall_call['waiting_duration_mean'], 
    data_mixed_internall_call['waiting_duration_mean'], equal_var = False)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Нулевая гипотеза не отвергнута, есть вероятность, что среднее время ожидания при входящих внешних вызовах одинаковое в группах `in` и `mixed`.

Построим гистограмму распределения  среднего времени ожидания при входящих внешних вызовах в группах `in` и `mixed`.

```python
data_in_internall_call['waiting_duration_mean'].plot(kind='hist', bins=15, 
                              grid=True, alpha=0.7, label='in', fontsize=15) 
data_mixed_internall_call['waiting_duration_mean'].plot(kind='hist', bins=15, 
                              grid=True, alpha=0.7, label='mixed', fontsize=15)
plt.xlabel('Средняя время ожидания', fontsize=15)
plt.ylabel('Частота', fontsize=15)
plt.legend()
plt.title('Распределение среднего времени ожидания при входящих внешних вызовах в группах in и mixed', fontsize=15);
```

По критерию "долгое время ожидания ответа при входящих внешних вызовах" выявлено 73 неэффективных оператора:  51 оператор в категории "входящие операторы" и 22 оператора в категории "многофункциональные операторы".
По критерию "мало исходящих вывозов" выявлено 755 неэффективных операторов: 589 операторов в категории "исходящие операторы" и 166 - в категории "многофункциональные операторы".

Мы рассмотрели две гипотезы:
* Среднее время ожидания при входящих внешних вызовах в группах `in` и `mixed` одинаковое;
* Среднее количество пропущенных входящих вызовов в день в группах `in` и `mixed` одинаковое..

При проверке гипотез установили, что нет статистического отличия среднего время ожидания при входящих внешних вызовах и среднего количествва пропущенных входящих вызовов в день в группах `in` и `mixed`.



**Рекомендации**

Рекомендовано провайдеру внедрить новый сервис для определения неэффективных операторов.
Это поможет привлечь новых клиентов, и мотивирует остаться именно с этим
провайдером действующих клиентов.
Для определения неэффективности использовать следующие критерии:
* процент пропущенных звонков не должен превышать порог в 3%;
* время ожидания ответа при входящих внешних вызовах должен быть не более 28 секунд;
* совершает не менее 144 звонков в день — в случае тех колл-центров, что специализируются
на обзвонах, а не обработке входящих звонков.
